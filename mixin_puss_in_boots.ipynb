```python

class BaseCharacter:
    def __init__(self, name):
        self.name = name


class BaseHuman:
    def __init__(self):
        self.is_human = True
```

```python

class MixinSpeakable:
    def speakness(self):
        print(f'{self.name} говорит: {self.phrase}')


class MixinActionable:
    def perfom_action(self):
        self.energy = int(self.energy) - 10
        print(f'Силы жить {self.name}: {self.energy}')


class MixinCollectible:
    is_collectible = True

    def collect(self):
        print(f'{self.name} добавлен в коллекцию!')


class MixinPoseable:
    is_poseable = True

    def change_pose(self, pose):
        print(f'{self.name} принимает позу: {pose}')


class MixinCostumeWearable:
    def wear_costume(self, costume):
        print(f'{self.name} надевает костюм: {costume}')


class MixinAnimated:
    is_animated = True

    def animate(self):
        print(f'{self.name} оживает!')


class MixinFunny:
    def make_laugh(self):
        print(f'{self.name} шутит и все смеются!')



class BaseInActionCharacter(MixinSpeakable, MixinActionable, BaseCharacter):
    def __init__(self, name):
        BaseCharacter.__init__(self, name)


class BaseFunkoPop(MixinCollectible, MixinPoseable, BaseCharacter):
    def __init__(self, name):
        BaseCharacter.__init__(self, name)

    def display(self):
        print(f'Фигурка {self.name} стоит на полке')


class BaseCosplayer(MixinCostumeWearable, MixinFunny, BaseCharacter, BaseHuman):
    def __init__(self, name):
        BaseCharacter.__init__(self, name)
        BaseHuman.__init__(self)
```

```python


class Shrek(BaseCharacter):
    def __init__(self):
        BaseCharacter.__init__(self, 'Шрек')
        self.phrase = 'Это моё болото!'


class ShrekInAction(BaseInActionCharacter):
    def __init__(self):
        BaseInActionCharacter.__init__(self, 'Шрек')

        self.phrase = 'Это моё болото!'
        self.energy = 100


class ShrekFunkoPop(BaseFunkoPop):

    def __init__(self):

        BaseFunkoPop.__init__(self, 'Шрек')
        print('Полка превращается в болото')



class ShrekCosplayer(BaseCosplayer):

    def __init__(self):

        super().__init__('Шрек')

        self.costume = 'Зеленый костюм Шрека с ушами на ободке'
```
