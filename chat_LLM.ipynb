```python
import os
from dotenv import load_dotenv

load_dotenv()

api_key = os.environ.get("API_KEY")
```

```python
import os
from openai import OpenAI as oai
from rich.console import Console
from rich.panel import Panel as pan


url = "https://foundation-models.api.cloud.ru/v1"

MODEL_NAME = "t-tech/T-lite-it-1.0"

client = oai(
    api_key=api_key,
    base_url=url
)

console = Console()
history = []

while True:
    user_input = input("Ты: ").strip()

    if user_input == "/exit":
        console.print("[bold red]До встречи![/bold red]")
        break

    if user_input == "/clear":
        history = [msg for msg in history if msg["role"] == "system"]
        console.print("[yellow]История очищена[/yellow]")
        continue

    if user_input.startswith("/system "):
        sys_text = user_input[8:]
        history = [msg for msg in history if msg["role"] != "system"]
        history.insert(0, {"role": "system", "content": sys_text})
        console.print("[hot_pink]Системный промпт установлен[/hot_pink]")
        continue

    if not user_input:
        continue

    console.print(pan(user_input, title="[white]Вы[/white]", border_style="white"))

    history.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model=MODEL_NAME,
        max_tokens=2500,
        temperature=0.5,
        presence_penalty=0,
        top_p=0.95,
        messages=history
    )


    answer = response.choices[0].message.content
    history.append({"role": "assistant", "content": answer})

    console.print(pan(answer, title="[bright_green]Модель[/bright_green]", border_style="green"))
```
